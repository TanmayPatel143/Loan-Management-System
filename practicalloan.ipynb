```python
######### using the LDA

from sklearn.metrics import classification_report,cohen_kappa_score,accuracy_score,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

data=pd.read_csv('cleanedloan.csv')
# data.iloc[:10,5:]

x=data.iloc[:,[4,6,8,11]].values
y=data.iloc[:,28].astype(int)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

lda=LDA(n_components=None)
xtrain=lda.fit_transform(x_train,y_train)
xtest=lda.transform(x_test)

print('original feature:',x.shape[1])
print("reduced number of Features:",xtrain.shape[1])
print("Variance Ratio:",lda.explained_variance_ratio_)

    
```

```python
#prediction using the LDA

knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)

classify_report(y_test,y_pred_knn)
```

```python
#using the PCA
from sklearn.decomposition import PCA

pca=PCA(n_components=0.99)
print('original feature:',xtrain.shape[1])
print('original test feature:',xtest.shape[1])
x_train=pca.fit_transform(x_train)
x_test=pca.transform(x_test)

print('reduced number of feature:',xtrain.shape[1])
print('reduced number of testing feature:',xtest.shape[1])
```

```python
#prediction using the PCA

knn=KNeighborsClassifier(n_neighbors=10)
knn.fit(x_train,y_train)
y_pred_knn2=knn.predict(x_test)
classify_report(y_test,y_pred_knn2)
```

```python
#using the NMF
from sklearn.decomposition import NMF

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 0) 

nmf=NMF(n_components=2,init='random',random_state=0)
xtrain=nmf.fit_transform(xtrain)
xtest=nmf.transform(xtest)

print('original number of features:',xtrain.shape[1])
print('reduced number of features:',xtrain.shape[1])
print("reduced number of testing features:",xtest.shape[1])
```

```python
#prediction using the NMF

knn = KNeighborsClassifier(n_neighbors=10)
#Fit a model
knn.fit(xtrain,ytrain)
# Perform predictions
y_pred_knn = knn.predict(xtest)
#Classification report
classify_report(ytest,y_pred_knn)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler

# Load dataset
# data = pd.read_csv("your_file.csv")  # Uncomment if needed

# Reduce dataset size (random sample of 5000 rows)
data_sample = data.sample(n=5000, random_state=42)

# Selecting features
x = data_sample.iloc[:, [29,31,32,38,41,42,47,48]].values  # Selecting specific columns

# Scale data to improve clustering
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

# Train Hierarchical Clustering (Agglomerative)
hc = AgglomerativeClustering(n_clusters=3, linkage='ward')
y_pred = hc.fit_predict(x_scaled)

# Visualizing the clusters
plt.figure(figsize=(10, 8))
plt.scatter(x_scaled[y_pred == 0, 0], x_scaled[y_pred == 0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(x_scaled[y_pred == 1, 0], x_scaled[y_pred == 1, 1], s=100, c='green', label='Cluster 2')
plt.scatter(x_scaled[y_pred == 2, 0], x_scaled[y_pred == 2, 1], s=100, c='red', label='Cluster 3')

plt.title('Clusters of Customers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
